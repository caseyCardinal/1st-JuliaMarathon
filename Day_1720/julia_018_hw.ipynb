# DataFrames.jl 介紹 (二): Joins 與 Split-Apply-Combine Strategy

## Day 018 作業1：The Split-Apply-Combine Strategy

同 Day 017，請載入 2020/4/2 COVID-19 Daily Report 資料集，計算每個國家/地區的累計確診人數，並列出所有資料筆數。

```julia
using DataFrames, CSV
```

```julia
df = CSV.read("04-02-2020.csv")
```

計算計算每個國家/地區的累計確診人數

【提示】可使用 groupby 相關的函式

```julia
gd = groupby(df, :Country_Region)
```

```julia
using Printf
```

```julia
for i in 1:length(gd)
    gdi=gd[i]
    cfrmNum=sum(gdi[!, :Confirmed])
    ctryRgn=gdi[1, :Country_Region]
    @printf("Country_Region=%-32s \tConfirmed=%d\n", ctryRgn, cfrmNum)
end
```

## 作業2：請列出美國各區域或事件 (依 Province_State 欄) 的累積確診、死亡、康復人數。

【提示】使用 Split-Apply-Combine Strategy 時，有幾種不同的函式可以使用，請參照 `DataFrames.jl` 官方文件 [https://juliadata.github.io/DataFrames.jl/stable/lib/functions/#DataFrames.groupby](https://juliadata.github.io/DataFrames.jl/stable/lib/functions/#DataFrames.groupby)

```julia
dfUS=df[ df[!, :Country_Region].=="US", :]
gdUS=groupby(dfUS, :Province_State)
```

```julia
for i in 1:length(gdUS)
    gdi=gdUS[i]
    cfrmNum=sum(gdi[!, :Confirmed])
    ctryRgn=gdi[1, :Province_State]
    @printf("State=%-24s\t%6d Confirmed\n", ctryRgn, cfrmNum)
end
```

```julia
for i in 1:length(gdUS)
    gdi=gdUS[i]
    ctryRgn=gdi[1, :Province_State]
    cfrmNum=sum(gdi[!, :Confirmed])
    dNum=sum(gdi[!, :Deaths])
    rcvNum=sum(gdi[!, :Recovered])

    @printf("State=%-24s\t%6d Confirmed\t %6d dead\t%6d Recovered\n", ctryRgn, cfrmNum, dNum, rcvNum)
end
```
